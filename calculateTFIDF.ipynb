```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
pd.set_option("display.max_rows", 600)
from pathlib import Path  
from nltk.corpus import stopwords
import json
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
lmtzr = WordNetLemmatizer()
#word_tokenize(text, language='french', preserve_line=False)
```

```python
pathCorpus = "/home/luis/Documents/french-docs/"
#Reading Jason with all information about corpus
with open(pathCorpus+"french-docs.json", "r") as file:
    data = json.load(file)
#Reading name's file of news
articles=[]
for key in data :
    articles.append(pathCorpus+data[key]['document_path'])
```

```python
# Interface lemma tokenizer from nltk with sklearn
class LemmaTokenizer:
    ignore_tokens = [',', '.', ';', ':', '"', '``', "''", '`']
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t) for t in word_tokenize(doc) if t not in self.ignore_tokens]
```

```python
# Lemmatize the stop words
tokenizer=LemmaTokenizer()
stop_words = set(stopwords.words('french')) 
token_stop = tokenizer(' '.join(stop_words))
```

```python
text_titles = [Path(text).stem for text in articles]
```

```python
text_titles
```

```python
tfidf_vectorizer = TfidfVectorizer(input='filename', stop_words=stopwords.words("french"), tokenizer=word_tokenize('french')
```

```python
tfidf_vector = tfidf_vectorizer.fit_transform(articles)
```

```python
tfidf_df = pd.DataFrame(tfidf_vector.toarray(), index=text_titles, columns=tfidf_vectorizer.get_feature_names_out())
```
